# LABEL ENCODING

converting categorical class into numbers

#METHOD 1

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
print("Before Label ENcoder ",count(df.sex))
df.sex=le.fit_transform(df.sex)     #encode to numerical values
print("After Label ENcoder ",count(df.sex))
```

METHOD 2

```python
print("Before Label ENcoder ",count(df.alive))
df['alive']=df['alive'].replace(['yes','no'],[1,0])    #encode to numerical values
print("After Label ENcoder ",count(df.alive))
```

```python
print("Before Label ENcoder ",count(df.who))
df['who']=df['who'].replace(['woman','man','child'],[1,2,3])    #encode to numerical values
print("After Label ENcoder ",count(df.who))
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
print("Before Label ENcoder ",count(df.alone))
df.alone=le.fit_transform(df.alone)     #encode to numerical values
print("After Label ENcoder ",count(df.alone))
```
